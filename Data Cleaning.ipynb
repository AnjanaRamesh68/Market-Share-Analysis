```python
months = {'may_22':may_22,'jun_22':jun_22,'jul_22':jul_22,'aug_22':aug_22,'sep_22':sep_22,'oct_22':oct_22,'nov_22':nov_22,'dec_22':dec_22,'jan_23':jan_23,'feb_23':feb_23,'mar_23':mar_23}
type_1 = {}
type_2 = {}
for i,j in months.items():
    a = j.columns[1]
    if 'Unnamed' in a:
        type_1[i] = j
    else:
        type_2[i] = j
        
```

```python
def bank_t(x):
    bank_type = []
    bt = ''
    for i in list(x['b']):
        if isinstance(i,str):
#             print(i)
            bt = i
            bank_type.append(bt)
        else:
            bank_type.append(bt)
    return bank_type
```

```python
for name,df in type_1.items():
    new_df = df
    print(name)
    a = new_df[new_df['Unnamed: 1'] == 'Public Sector Banks'].index[0]
    b = new_df[new_df['Unnamed: 1'] == 'Payment Banks'].index[0]
    new_df = new_df.fillna('TBh')
    new_df.columns = new_df.iloc[2]+new_df.iloc[3]+new_df.iloc[4]+new_df.iloc[5]
    del_rows = [i for i in range(0,int(a))]
    del_rows2 = [i for i in range(int(b),len(new_df))]

    new_df = new_df.drop(del_rows)
    new_df = new_df.drop(del_rows2)
    new_df.columns = new_df.columns.str.replace(r'\s+', ' ')


    try:
        new_df = new_df.drop(["Number - Outstanding (as on month end)ATMTBhOn-site",
           'TBhTBhTBhOff-site', 'TBhPoSTBhTBh', 'TBhMicro ATMsTBhTBh',
           'TBhBharat QR CodesTBhTBh','TBhUPI QR CodesTBhTBh','TBhCash WithdrawalAt ATMVolume (in actuals)',
           "TBhCash WithdrawalATMVolume (in actuals)","TBhTBhPoSVolume (in actuals)"],axis = 1)

    except:
          new_df = new_df.drop(['Number - Outstanding (as on month end)ATMs & CRMsTBhOn-site','TBhTBhTBhOff-site', 'TBhPoSTBhTBh', 'TBhMicro ATMsTBhTBh',
           'TBhBharat QR CodesTBhTBh', 'TBhUPI QR CodesTBhTBh','TBhCash WithdrawalAt ATMVolume (in actuals)',
           "TBhCash WithdrawalATMVolume (in actuals)","TBhTBhPoSVolume (in actuals)"],axis = 1)

    new_df.columns = ['a','b','Bank_name','Credit_cards','Debit_Cards','cc_pos_vol','cc_pos_val','cc_online_vol','cc_online_val','cc_other_vol','cc_other_val','remov','dc_pos_vol','dc_pos_val','dc_online_vol','dc_online_val','dc_other_vol','dc_other_val','remov1','remov2']
    new_df['bank_type'] = bank_t(new_df)
    new_df.drop(['a','b','remov','remov1','remov2'],axis = 1,inplace = True)
    new_df = new_df[new_df['Bank_name']!='TBh']

    new_df['cc_transactions'] = new_df['cc_pos_vol'] + new_df['cc_online_vol'] + new_df['cc_other_vol']
    new_df['dc_transactions'] = new_df['dc_pos_vol'] + new_df['dc_online_vol'] + new_df['dc_other_vol']
    new_df['cc_inr_value'] = ((new_df['cc_pos_val']*1000) + (new_df['cc_online_val']*1000) + (new_df['cc_other_val']*1000))
    new_df['dc_inr_value'] = ((new_df['dc_pos_val']*1000) + (new_df['dc_online_val']*1000) + (new_df['dc_other_val']*1000))

    new_df = new_df.reset_index()
    new_df.drop('index',axis = 1,inplace  = True)
    new_df = new_df[['Bank_name','bank_type','Credit_cards','Debit_Cards','cc_transactions','cc_inr_value','dc_transactions','dc_inr_value']]
    new_df.to_excel(f'D:\Projects\HDFC\CleanedExcel\ {name}.xlsx',index = False)
```
